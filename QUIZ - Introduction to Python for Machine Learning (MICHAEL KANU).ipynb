```python
import pandas as pd

# Select the columns of interest
columns_of_interest = ['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']

# Calculate the correlation matrix
correlation_matrix = df[columns_of_interest].corr()

# Exclude the 'Element Code' row since we are interested in its correlation with years
element_code_correlation = correlation_matrix.loc['Element Code']

# Find the year with the minimum correlation with 'Element Code'
least_correlated_year = element_code_correlation.idxmin()

print("Year with the least correlation with 'Element Code':", least_correlated_year)
print("Correlation coefficient:", element_code_correlation[least_correlated_year])
```

```python
import pandas as pd

# Select columns 'Y2017' and 'Area'
selected_columns = ['Area', 'Y2017']
selected_data = df[selected_columns]

# Group by 'Area' and calculate the sum of 'Y2017' for each area
sum_by_area = selected_data.groupby('Area')['Y2017'].sum()

# Find the area with the highest sum
area_with_highest_sum = sum_by_area.idxmax()
highest_sum = sum_by_area.max()

print("Area with the highest sum in 2017:", area_with_highest_sum)
print("Sum in 2017:", highest_sum)
```

```python
import pandas as pd

# Select columns 'Element' and 'Y2017'
selected_columns = ['Element', 'Y2017']
selected_data = df[selected_columns]

# Filter the data to include only rows where the 'Element' is 'Processing'
processing_data = selected_data[selected_data['Element'] == 'Processing']

# Group by 'Element' and calculate the sum of 'Y2017' for the 'Processing' element
sum_processing_2017 = processing_data.groupby('Element')['Y2017'].sum()

print("Total sum of Processing in 2017:", sum_processing_2017.values[0])
```

```python
# Total number of missing values in 2014
missing_values_2014 = df['Y2014'].isnull().sum()

# Total number of entries in 2014
total_entries_2014 = df['Y2014'].shape[0]

# Percentage of missing values in 2014
percentage_missing_2014 = (missing_values_2014 / total_entries_2014) * 100

# Round to 3 decimal places
percentage_missing_2014 = round(percentage_missing_2014, 3)

print("Total number of missing values in 2014:", missing_values_2014)
print("Percentage of missing values in 2014:", percentage_missing_2014)
```
